```python
# Import libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset('iris')
```

```python
# 1. Understand the dataset
print("First 5 rows of the dataset:")
print(df.head())
print("\nDataset Info (columns, data types, missing values):")
print(df.info())
print("\nSummary Statistics (numerical columns):")
print(df.describe())
```

```python
# 2. Check for missing values
print("\nMissing Values:")
print(df.isnull().sum())
```

```python
# 3. Univariate Analysis
# Histogram for sepal_length
plt.figure(figsize=(8, 6))
df['sepal_length'].hist(bins=20, color='#1f77b4')
plt.title('Histogram of Sepal Length')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Frequency')
plt.show()
```

```python
# Boxplot for petal_length
plt.figure(figsize=(8, 6))
sns.boxplot(x='petal_length', data=df, color='#ff7f0e')
plt.title('Boxplot of Petal Length')
plt.show()
```

```python
# Count plot for species (categorical)
plt.figure(figsize=(8, 6))
sns.countplot(x='species', data=df, palette='Set2')
plt.title('Count of Each Species')
plt.show()
```

```python
# 4. Bivariate Analysis
# Scatter plot: sepal_length vs sepal_width, colored by species
plt.figure(figsize=(8, 6))
sns.scatterplot(x='sepal_length', y='sepal_width', hue='species', size='species', data=df)
plt.title('Sepal Length vs Sepal Width by Species')
plt.show()
```

```python
# Pairplot: relationships across all numerical variables
sns.pairplot(df, hue='species', palette='Set2')
plt.show()
```

```python
# Correlation matrix (only numerical columns)
print("\nCorrelation Matrix:")
print(df.corr(numeric_only=True))
```

```python
# Heatmap of correlations
plt.figure(figsize=(8, 6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm', center=0)
plt.title('Correlation Heatmap')
plt.show()
```

```python
# 5. Grouped Analysis
print("\nMean of each feature by Species:")
print(df.groupby('species').mean())
```
